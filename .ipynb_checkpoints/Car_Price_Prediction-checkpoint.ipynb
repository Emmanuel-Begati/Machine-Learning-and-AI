I have began the features engineering and data preprocessing of the data. I see that there are some features like 'symboling', 'CarName', and 'car_ID', that do not need to be used by me and the model... so i am going to drop them


```python
df['CarName'].unique()
```

We are going to only use the company name so that we are able to use the carName as well

```python
df['brand'] = df.CarName.str.split(' ').str.get(0).str.lower()
```

```python
df.brand.unique()
```

```python
df['brand'] = df['brand'].replace(['maxda'], 'mazda')
df['brand'] = df['brand'].replace(['porcshce'], 'porsche')
df['brand'] = df['brand'].replace(['toyouta'], 'toyota')
df['brand'] = df['brand'].replace(['vokswagen', 'vw'], 'volkswagen')
```

```python
df.head()
```

```python
df.drop(['car_ID', 'symboling', 'CarName'], axis=1, inplace=True)
```
