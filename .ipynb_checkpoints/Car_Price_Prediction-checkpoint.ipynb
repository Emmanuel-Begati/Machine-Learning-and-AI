I have began the features engineering and data preprocessing of the data. I see that there are some features like 'symboling', 'CarName', and 'car_ID', that do not need to be used by me and the model... so i am going to drop them


```python
df['CarName'].unique()
```

We are going to only use the company name so that we are able to use the carName as well

```python
df['brand'] = df.CarName.str.split(' ').str.get(0).str.lower()
```

```python
df.brand.unique()
```

```python
df['brand'] = df['brand'].replace(['maxda'], 'mazda')
df['brand'] = df['brand'].replace(['porcshce'], 'porsche')
df['brand'] = df['brand'].replace(['toyouta'], 'toyota')
df['brand'] = df['brand'].replace(['vokswagen', 'vw'], 'volkswagen')
```

```python
df.head()
```

```python
df.drop(['car_ID', 'symboling', 'CarName'], axis=1, inplace=True)
```

#                                             Features Engineering

### Drivewheel column

```python
df.drivewheel.unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
encoder.fit(df[['drivewheel']])
```

```python
drive_wheel_encoded = encoder.transform(df[['drivewheel']])
```

```python
drive_wheel_encoded
```

```python
drive_wheel_encoded = drive_wheel_encoded.toarray()
```

```python
drive_wheel_encoded
```

```python
new_drive_wheel = pd.DataFrame(drive_wheel_encoded, columns=encoder.get_feature_names_out(['drivewheel']))
print(new_drive_wheel)#new drivewheel value
```

### Fueltype Column

```python
df.fueltype.value_counts()
```

```python
to_encode_fueltype = pd.DataFrame(df['fueltype'])
binary_encoder = BinaryEncoder()
#we are using binary encoder because we see that gas has a lot more values
#than diesel. Binary encoder is a lot more effecient to use in such cases than One Hot encoder
```

```python
binary_encoder.fit(to_encode_fueltype)
```

```python
new_fuel_type = binary_encoder.transform(to_encode_fueltype)
```

```python
new_fuel_type #new fueltype value
```

```python
new_fuel_type_columns = ['fueltype_diesel', 'fueltype_gas']
new_fuel_type.columns = new_fuel_type_columns
new_fuel_type
```

The change has been made

### Doornumber Column

```python
df.doornumber.value_counts()
```

```python
binary_encoder.fit(df['doornumber'])
new_doornumber = binary_encoder.transform(df['doornumber'])
```

```python
door_mapping = {'two': 2, 'four': 4}

# Applying the ordinal encoding
df['doornumber'] = df['doornumber'].map(door_mapping)
```
